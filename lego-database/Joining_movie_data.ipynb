```python
#Finding all Star Wars in movies, parent_name was in lego_out
pattern = re.compile(".*Star Wars.*")
movie_set['star_wars'] = [True if pattern.match(str(item)) else False for item in movie_set['Title']]
starwars_movies = movie_set[movie_set['star_wars']==True]
starwars_movies
```

```python
#plotting star wars set releases by year
sw_years = starwars_legos['year'].astype(int)
sw_graph = sw_years.groupby(starwars_legos['year'].astype(int)).count().plot(kind="bar", title ="Star Wars Set Releases by Year", fontsize=12)
```
