```python
import csv
import pandas as pd
```

```python
#need to see what the dataset looks like
df = pd.read_csv("lego_out.csv")
df.head(10)
```

```python
#Need the year and movie names to match back to the movie dataset
#Year is obviously in column year but movie titles could be a mixture of set_name, theme_name, and parent_name

smaller_df = df[['year','set_name','theme_name','parent_name']]

smaller_df.head()
```

```python
#have to get unique values of those different columns

grouped_df = smaller_df.drop_duplicates()

grouped_df.head()
```

```python
parentname_col = grouped_df['parent_name'].unique()
print(parentname_col[:20])
#movies here are 'Star Wars' and 'Harry Potter'
#'Super Heroes' probably has to do with super hero movies
```

```python
themename_col = grouped_df['theme_name'].unique()
print(themename_col)
#franchises here are 'Spider-Man', 'The LEGO Movie', 'Iron Man', 'Teenage Mutant Ninja Turtles', 'X-Men'
```

```python
setname_col = grouped_df['set_name'].unique()
print(setname_col[:50])
#These don't look like movie names so we can eliminate this column
```

```python
#set_name isn't needed because it doesn't have movie names in it
fourcol_df = grouped_df.drop(['set_name'], axis = 1)
fourcol_df.drop_duplicates()
```

```python
#Should join theme name and parent name so that when checking against the movies database, the movie titles can check both columns

fourcol_df['full_name'] = fourcol_df['parent_name'].fillna('') + "," + fourcol_df['theme_name'].fillna('')
fourcol_df.drop_duplicates()
```

```python
#Now all we need is the combined row to check against movie titles. Then we can export to a CSV and join it back to the movie data in another notebook

final_df = fourcol_df.drop(['theme_name','parent_name'], axis = 1)
final_df.drop_duplicates()
```

```python
final_unique = final_df.drop_duplicates()
```

```python
#make it a csv so I can work on both this and the movies dataset together in a different notebook
final_unique.to_csv('out.csv', sep=',', index = False)
```
