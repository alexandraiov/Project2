```python
with open('oneyear_test.csv', 'w') as csvfile:
    filewriter = csv.writer(csvfile, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    filewriter.writerow(['Year', 'Rank', 'Title'])
    for x in range(0,50):
        this_movie = movie_containers1[x]
        filewriter.writerow(['2017',this_movie.td.text,this_movie.b.a.text]) 
    for x in range(0,50):
        this_movie = movie_containers2[x]
        filewriter.writerow(['2017',this_movie.td.text,this_movie.b.a.text]) 
```
